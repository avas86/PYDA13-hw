### Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
ratings.info()
```

```python
movies = pd.read_csv('ml-latest-small/movies.csv')
movies.head()
```

```python
movies.info()
```

```python
join = ratings.merge(movies, on='movieId', how='left')
join.head()
```

```python
best_movies = join[join['rating'] == 5]
best_movies['title'].value_counts()
```

```python
result = best_movies['title'].value_counts().index.values[0]
print(f'Фильму {result} выставлено больше всего оценок 5.0')
```

### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
data.head()
```

```python
data.info()
```

```python
# последовательное применение фильтров к dataset'у
country = data[(data['country']=='Estonia') | (data['country']=='Latvia') | (data['country']=='Lithuania')]
cat = country[(country['category'] == 4) | (country['category'] == 12) | (country['category'] == 21)]
year = cat[(cat['year'] >= 2005) & (cat['year'] <= 2010)]
quantity = year[year['quantity'] >= 0]
count_total = quantity['quantity'].sum()
print(f'Суммарное потребление стран Прибалтики (кат. 4, 12 и 21) \nза период с 2005 по 2010год: {count_total}')
```

```python
# все фильтры в единой переменной
filters = data[((data['country']=='Estonia') | (data['country']=='Latvia') | (data['country']=='Lithuania')) & \
               ((data['category'] == 4) | (data['category'] == 12) | (data['category'] == 21)) & \
               ((data['year'] >= 2005) & (data['year'] <= 2010)) & \
               (data['quantity'] >= 0)]
count_total = filters['quantity'].sum()
print(f'Суммарное потребление стран Прибалтики (кат. 4, 12 и 21) \nза период с 2005 по 2010год: {count_total}')
```
