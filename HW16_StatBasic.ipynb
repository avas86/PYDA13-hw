### Задание 2. Работа с выбросами

В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

```python
q1 = df['Pulse'].quantile(0.25)
q3 = df['Pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['Pulse'].between(lower_bound, upper_bound, inclusive=True)].sort_values('Pulse')
remove_outliers.head()
```

```python
q1 = df['Pulse'].quantile(0.25)
q3 = df['Pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
outliers = df[(df['Pulse'] >= upper_bound) | (df['Pulse'] <= lower_bound)].sort_values('Pulse')
outliers
```

### Задание 3. Работа с пропусками

Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе работы с пропусками по каждому столбцу, сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
for col in df:
    missing = df[col].isnull().mean()
    print(f'{col} - {missing :.0%}')
```

```python
print(df.Pulse.mean())
print(df.Pulse.median())
print(df.Pulse.mode()[0])
print(df.Pulse.std())
print(df.Pulse.var())
```

оценим случаи, где пропущена температура

```python
no_rect = df.loc[(df['RectalTemperature'].isnull())]
no_rect.describe()
```

```python
for i in no_rect.columns:
    print(f'мода {i} - {df[i].mode()[0]}')
```

Как можно видеть показатели здоровья наиболее оптимальны, если рассчитаны по моде.

Дополним пропущенные значения температуры модой

```python
df['RectalTemperature'].fillna(df['RectalTemperature'].mode()[0], inplace=True)
df['RectalTemperature'].fillna(df['RectalTemperature'].mode()[0]).isna().sum()
```

так как количество пропущенных значений пульса незначительно, а полезность частоты дыхания малозначительна из за больших колебаний, целесооборазно заполнить недостающие значения медианой 

```python
df['Pulse'].fillna(df['Pulse'].median(), inplace=True)
df['Pulse'].fillna(df['Pulse'].median()).isna().sum()
```

```python
df['RespiratoryRate'].fillna(df['RespiratoryRate'].median(), inplace=True)
df['RespiratoryRate'].fillna(df['RespiratoryRate'].median()).isna().sum()
```
